```python
import pandas as pd
import numpy as np
from glob import glob
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_palette("colorblind")
```

```python
main_results_loc = 'results/'
noise_sigma = 0.1
```

```python
results_df = {}

results_df['sparse'] = pd.read_csv(main_results_loc + 'simulation_results_sparse.txt')
results_df['graph'] = pd.read_csv(main_results_loc +  'simulation_results_graph.txt')
```

The above fixes sparsity $s = 0.25$

```python
results_df['sparse']
```

```python
table_list = []

for simulation_type in ['sparse', 'graph']:
    print(simulation_type)
    for method in ['SCCA', 'GN-SCCA']:
        curr_row = []
        temp_df = results_df[simulation_type]
        temp_df = temp_df[temp_df['noise_sigma']== noise_sigma]
        temp_df = temp_df[temp_df['dim']==5]
        temp_df['avg_var'] = temp_df['sum_var']/5
        temp_df = temp_df[temp_df['method']==method]
        
        for deflation_type in ['HD', 'PD', 'OPD']:
            new_df = temp_df[temp_df['deflation']==deflation_type]
#             print(new_df)
            curr_row.append(str("{:.2f}".format(new_df['avg_var'].mean()*100)) + ' $\pm$ ' + 
                            str("{:.2f}".format(new_df['avg_var'].std()*100)))
            if method == 'GCCA':
                method = 'GN-SCCA-PG'
        print('& ' + method + ' & ' + ' & '.join(curr_row) + ' \\\\')
    print('\n')
```
