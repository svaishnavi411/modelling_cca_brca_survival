```python
curr_imaging_features = imaging_features
curr_gene_names = gene_names
```

```python
num_features = len(curr_imaging_features) 
node_labels = [x.split("_Histogram_")[0] for x in curr_imaging_features] + ["Image embedding " + str(i) for i in range(50)]
link_source = list(np.arange(len(curr_imaging_features))) #+
#                len(imaging_features) + np.arange(len(gene_names)))*50
 
link_target = []
for j in range(50):
    link_target.extend([num_features + j]*50)

link_value = np.reshape((np.concatenate((V), axis=0)).T, (-1, 1))
link_value[abs(link_value)<0.05] = 0


fig = go.Figure(data=[go.Sankey(
    valueformat = ".0f",
    valuesuffix = "TWh",
    # Define nodes
    node = dict(
      pad = 15,
      thickness = 15,
      line = dict(color = "black", width = 0.5),
      label =  node_labels,
#       color =  data['data'][0]['node']['color']
    ),
    # Add links
    link = dict(
      source =  link_source,
      target =  link_target,
      value =  link_value,
#       label =  data['data'][0]['link']['label'],
#       color =  data['data'][0]['link']['color']
))])

fig.update_layout(#title_text="Importance of imaging features in embeddings",
    width=500,
    height=400,
    font_size=8)

fig.write_image("1imaging_features_importance.png", scale=5)
```
