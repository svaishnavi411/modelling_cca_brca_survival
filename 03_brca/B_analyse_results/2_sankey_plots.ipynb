```python
num_features = len(curr_imaging_features) 

link_source = list(np.arange(len(curr_imaging_features))) 
link_target = []
for j in range(50):
    link_target.extend([num_features + j+1]*50)

link_value = np.abs(np.reshape((np.concatenate((V), axis=0)), (-1, 1)))
# link_value[np.abs(link_value)<0.05] = 0

node_labels = [x.split("_Histogram_")[0] for x in curr_imaging_features] + ["Image embedding " + str(i+1) for i in range(50)]
node_dict =  dict(
      pad = 15,
      thickness = 15,
      line = dict(color = "black", width = 0.5),
      label =  node_labels,
    #   x = [0.01]*len(curr_imaging_features) + [0.99]*50,
    #   y = [np.arange(len(curr_imaging_features))/len(curr_imaging_features)*25] + [np.arange(50)/50*255]
    )
fig = go.Figure(data=[go.Sankey(
    valueformat = ".0f",
    valuesuffix = "TWh",
    arrangement='snap',
    # Define nodes
    node = node_dict,
    # Add links
    link = dict(
      source =  link_source,
      target =  link_target,
      value =  link_value,
))])

fig.update_layout(#title_text="Importance of imaging features in embeddings",
    width=600,
    height=650,
    font_size=10)

fig.write_image("1imaging_features_importance.png", scale=5)
```
